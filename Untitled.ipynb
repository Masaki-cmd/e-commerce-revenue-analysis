## データクリーニングと整合性確認 (Data Cleaning and Consistency Check)

### Subtask:

**Reasoning**:
I will start by converting the 'Date' column to datetime objects, then impute missing values in 'Delivery_Days' with its median and 'Customer_Rating' with its mode. After that, I will replace any negative values in the specified monetary columns ('Unit_Price', 'Total_Price_Before_Discount', 'Discount_Amount', 'Price_After_Discount', 'Final_Total') with zero, as negative monetary values are illogical.



````python
import numpy as np

# 1. Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])
print("Converted 'Date' column to datetime.")

# 2. Handle missing values
# Impute 'Delivery_Days' with the median
median_delivery_days = df['Delivery_Days'].median()
df['Delivery_Days'].fillna(median_delivery_days, inplace=True)
print(f"Imputed missing 'Delivery_Days' with median: {median_delivery_days}.")

# Impute 'Customer_Rating' with the mode
mode_customer_rating = df['Customer_Rating'].mode()[0] # .mode() can return multiple if ties, so take first
df['Customer_Rating'].fillna(mode_customer_rating, inplace=True)
print(f"Imputed missing 'Customer_Rating' with mode: {mode_customer_rating}.")

# 3. Identify and handle negative values in numerical columns
negative_value_cols = ['Unit_Price', 'Total_Price_Before_Discount', 'Discount_Amount', 'Price_After_Discount', 'Final_Total']
for col in negative_value_cols:
    initial_neg_count = (df[col] < 0).sum()
    if initial_neg_count > 0:
        df[col] = df[col].apply(lambda x: max(0, x))
        print(f"Replaced {initial_neg_count} negative values in '{col}' with 0.")

print("\nDataFrame Information after initial cleaning:")
df.info()
print("\nDescriptive Statistics after initial cleaning:")
print(df.describe())
````

**Reasoning**:
The previous steps handled data type conversion, missing values, and negative values. Now, I will identify and remove any duplicate rows to ensure data integrity, which is a crucial part of data cleaning.



````python
import numpy as np

# 1. Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])
print("Converted 'Date' column to datetime.")

# 2. Handle missing values
# Impute 'Delivery_Days' with the median
median_delivery_days = df['Delivery_Days'].median()
df['Delivery_Days'].fillna(median_delivery_days, inplace=True)
print(f"Imputed missing 'Delivery_Days' with median: {median_delivery_days}.")

# Impute 'Customer_Rating' with the mode
mode_customer_rating = df['Customer_Rating'].mode()[0] # .mode() can return multiple if ties, so take first
df['Customer_Rating'].fillna(mode_customer_rating, inplace=True)
print(f"Imputed missing 'Customer_Rating' with mode: {mode_customer_rating}.")

# 3. Identify and handle negative values in numerical columns
negative_value_cols = ['Unit_Price', 'Total_Price_Before_Discount', 'Discount_Amount', 'Price_After_Discount', 'Final_Total']
for col in negative_value_cols:
    initial_neg_count = (df[col] < 0).sum()
    if initial_neg_count > 0:
        df[col] = df[col].apply(lambda x: max(0, x))
        print(f"Replaced {initial_neg_count} negative values in '{col}' with 0.")

# 4. Handle duplicate data
initial_rows = df.shape[0]
duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print(f"Removed {duplicate_rows} duplicate rows. DataFrame now has {df.shape[0]} rows.")
else:
    print("No duplicate rows found.")

print("\nDataFrame Information after initial cleaning:")
df.info()
print("\nDescriptive Statistics after initial cleaning:")
print(df.describe())
````

**Reasoning**:
I will cap the outliers in the numerical columns identified earlier, using the IQR method to replace values beyond the 1.5*IQR range with the calculated upper and lower bounds. This will help normalize the data distribution and prevent extreme values from unduly influencing subsequent analysis, while preserving the relative order of the data.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns for outlier detection
# Exclude 'Date' and 'Transaction_ID', 'Customer_ID' which are not directly numerical for this purpose
numerical_cols_for_outliers = [
    'Quantity',
    'Discount_Percentage',
    'Shipping_Cost',
    'Delivery_Days',
    'Customer_Rating',
    'Unit_Price',
    'Total_Price_Before_Discount',
    'Discount_Amount',
    'Price_After_Discount',
    'Final_Total'
]

print("\nVisualizing potential outliers using box plots:")

# Set up the matplotlib figure and axes
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols_for_outliers):
    plt.subplot(4, 3, i + 1) # Adjust subplot grid as needed
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel('') # Remove y-label as column name is title
plt.tight_layout()
plt.show()

print("\nConsideration for outliers: Based on the box plots, we can decide whether to remove, cap, or transform these outliers depending on their nature and impact on the analysis.")

# Outlier capping using IQR method
print("\nHandling outliers by capping using the IQR method:")
for col in numerical_cols_for_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    initial_outliers_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    if initial_outliers_count > 0:
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
        print(f"Capped {initial_outliers_count} outliers in column '{col}'.")
    else:
        print(f"No outliers detected for capping in column '{col}'.")

print("\nDataFrame Information after outlier capping:")
df.info()
print("\nDescriptive Statistics after outlier capping:")
print(df.describe())
````

````python
print("\n### 高評価顧客の属性分析\n")

# 1. 各Customer_IDごとの平均顧客評価を計算
average_customer_ratings_by_id = df.groupby('Customer_ID')['Customer_Rating'].mean().reset_index()

# 2. 平均評価が5.0の顧客を「高評価顧客」として特定
high_rated_customer_ids = average_customer_ratings_by_id[
    average_customer_ratings_by_id['Customer_Rating'] == 5.0
]['Customer_ID']

# 3. 高評価顧客の取引のみを抽出
df_high_rated_customers = df[df['Customer_ID'].isin(high_rated_customer_ids)]

# 4. 高評価顧客の購入行動を分析
print("#### 高評価顧客 (平均評価5.0) の購入行動概要:\n")

total_sales_high_rated = df_high_rated_customers['Final_Total'].sum()
total_quantity_high_rated = df_high_rated_customers['Quantity'].sum()

print(f"総売上: {total_sales_high_rated:,.2f}")
print(f"総販売数量: {total_quantity_high_rated:,}")

print("\n#### 高評価顧客のカテゴリー別購入割合:\n")
category_preference_high_rated = df_high_rated_customers['Category'].value_counts(normalize=True)
print(category_preference_high_rated)

print("\n#### 全顧客のカテゴリー別購入割合 (比較のため):\n")
category_preference_all = df['Category'].value_counts(normalize=True)
print(category_preference_all)

# Optionally, visualize category preference for high-rated customers
plt.figure(figsize=(10, 6))
sns.barplot(x=category_preference_high_rated.index, y=category_preference_high_rated.values, palette='viridis', hue=category_preference_high_rated.index, legend=False)
plt.title('高評価顧客のカテゴリー別購入割合')
plt.xlabel('カテゴリー')
plt.ylabel('購入割合')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 顧客評価が売上に与える間接的な影響の分析 ---
print("\n### 顧客評価が売上に与える間接的な影響の分析\n")

# 各Customer_IDごとの総売上と総販売数量を計算
customer_total_performance = df.groupby('Customer_ID').agg(
    total_final_total=('Final_Total', 'sum'),
    total_quantity=('Quantity', 'sum')
).reset_index()

# 平均顧客評価と顧客の総パフォーマンスを結合
customer_analysis_df = pd.merge(average_customer_ratings_by_id, customer_total_performance, on='Customer_ID')

print("#### 顧客ごとの平均評価と総パフォーマンスの記述統計:\n")
print(customer_analysis_df[['Customer_Rating', 'total_final_total', 'total_quantity']].describe())

# 顧客評価と総売上の関係を可視化
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_analysis_df, x='Customer_Rating', y='total_final_total', alpha=0.6, hue='Customer_Rating', palette='viridis')
plt.title('顧客の平均評価と総売上の関係')
plt.xlabel('顧客の平均評価')
plt.ylabel('総売上')
plt.grid(True)
plt.tight_layout()
plt.show()

# 顧客評価と総販売数量の関係を可視化
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_analysis_df, x='Customer_Rating', y='total_quantity', alpha=0.6, hue='Customer_Rating', palette='magma')
plt.title('顧客の平均評価と総販売数量の関係')
plt.xlabel('顧客の平均評価')
plt.ylabel('総販売数量')
plt.grid(True)
plt.tight_layout()
plt.show()

# ピアソン相関係数の計算
correlation_rating_total_sales = customer_analysis_df['Customer_Rating'].corr(customer_analysis_df['total_final_total'])
correlation_rating_total_quantity = customer_analysis_df['Customer_Rating'].corr(customer_analysis_df['total_quantity'])

print(f"\n顧客の平均評価と総売上の相関係数: {correlation_rating_total_sales:.2f}")
print(f"顧客の平均評価と総販売数量の相関係数: {correlation_rating_total_quantity:.2f}")
````

**Reasoning**:
To perform deeper customer segmentation as suggested in the 'Next steps' from the previous summary, I will segment customers based on their average `Customer_Rating` and `total_final_total` (as a proxy for CLV). This will create distinct customer groups, allowing for a more nuanced analysis of their characteristics and behaviors.



````python
print("\n### 顧客セグメンテーションの深化: 評価と支出による顧客グループ分析\n")

# Customer_Analysis_df はすでに存在し、Customer_IDごとの平均評価と総支出が含まれている
# customer_analysis_df = pd.merge(average_customer_ratings_by_id, customer_total_performance, on='Customer_ID')

# 顧客評価のセグメント化
def segment_rating(rating):
    if rating < 4.0:
        return 'Low_Rating'
    elif rating < 4.5:
        return 'Medium_Rating'
    else:
        return 'High_Rating'

customer_analysis_df['Rating_Segment'] = customer_analysis_df['Customer_Rating'].apply(segment_rating)

# 総支出（LTVプロキシ）のセグメント化
Q1_total_spend = customer_analysis_df['total_final_total'].quantile(0.25)
Q3_total_spend = customer_analysis_df['total_final_total'].quantile(0.75)

def segment_spend(total_spend):
    if total_spend < Q1_total_spend:
        return 'Low_Spender'
    elif total_spend < Q3_total_spend:
        return 'Medium_Spender'
    else:
        return 'High_Spender'

customer_analysis_df['Spend_Segment'] = customer_analysis_df['total_final_total'].apply(segment_spend)

print("#### 顧客評価セグメントと支出セグメントの組み合わせ分布:\n")
segment_counts = customer_analysis_df.groupby(['Rating_Segment', 'Spend_Segment']).size().unstack(fill_value=0)
print(segment_counts)

print("\n#### 各セグメントの平均顧客評価、平均総支出、平均販売数量:\n")
segment_summary = customer_analysis_df.groupby(['Rating_Segment', 'Spend_Segment']).agg(
    Avg_Customer_Rating=('Customer_Rating', 'mean'),
    Avg_Total_Spend=('total_final_total', 'mean'),
    Avg_Total_Quantity=('total_quantity', 'mean')
)
print(segment_summary)

# 可視化: 評価セグメントと支出セグメントの棒グラフ (数)
plt.figure(figsize=(12, 7))
segment_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('顧客評価セグメントと支出セグメントの組み合わせ分布')
plt.xlabel('顧客評価セグメント')
plt.ylabel('顧客数')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 可視化: 各セグメントの平均総支出
plt.figure(figsize=(12, 7))
segment_summary['Avg_Total_Spend'].unstack().plot(kind='bar', colormap='magma')
plt.title('顧客評価セグメントと支出セグメントごとの平均総支出')
plt.xlabel('顧客評価セグメント')
plt.ylabel('平均総支出')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

## 最終報告書：Eコマース売上データ分析に基づく収益最大化戦略

### はじめに

この報告書は、2023-2024年のeコマース売上データを詳細に分析し、収益最大化とコスト最適化のための具体的な戦略と提案をまとめたものです。データ抽出、クリーニング、仮説構築、詳細なデータ分析、可視化、および投資対効果（ROI）評価を経て得られた主要な知見に基づき、実行可能な施策を提言します。

### 1. 主要な分析結果と洞察

これまでの分析フェーズで得られた主要な発見と洞察を、以下の項目に分けてまとめます。

#### 1.1 製品カテゴリーごとの収益貢献度

*   **高収益カテゴリー（Electronics）** が総売上を牽引しており、その高単価戦略が成功していることが示唆されました。
*   **Accessories (アクセサリー)** と **Peripherals (周辺機器)** は販売数量が多いものの、総売上はElectronicsに比べて低い傾向にあり、単価の低さが影響していると考えられます。
*   **Networking (ネットワーク機器)** と **Storage (ストレージ)** は、総売上・販売数量ともに低く、改善の余地が大きいことが判明しました。

#### 1.2 割引戦略の最適化

*   取引の約50%に割引が適用されていますが、**割引率と販売数量、最終総額の間に強い正の相関は確認されませんでした**。回帰分析の結果も、統計的に有意な関係を示していません。
*   特に20%割引が適用された取引は、他の割引率と比較して平均最終総額が低い傾向にあり、高割引が必ずしも高収益に繋がっていない可能性が指摘されます。
*   割引率は0%, 5%, 10%, 15%, 20%に集中しており、段階的な割引戦略がとられているものの、その効果はカテゴリーによって異なると考えられます。例えば、Electronicsカテゴリーでは高額割引でも高い最終総額を維持しています。

#### 1.3 顧客評価と売上の相関関係

*   **顧客評価（Customer_Rating）と平均最終総額・平均販売数量の間には、直接的な強い相関は見られませんでした**。むしろ、最高評価（5.0）の顧客で平均最終総額がわずかに低いという逆説的な傾向が観察されました。
*   リピート顧客の平均顧客評価は全顧客とほぼ同じであり、リピート購入率との直接的な関連も薄いことが示唆されました。
*   製品ごとの平均顧客評価に基づいた「トップ評価製品」と「ボトム評価製品」の比較では、トップ評価製品群の平均売上はボトム評価製品群よりも低いという結果になりました。これは、必ずしも高評価製品が高売上につながるわけではないことを示しています。
*   顧客レベルでの平均評価と総支出（総売上）・総販売数量の相関係数も非常に弱く、顧客評価が売上に与える影響は間接的または遅延的である可能性が示唆されます。

#### 1.4 地域ごとの販売実績

*   **米国、ドイツ、カナダ**が総売上において上位を占める主要な収益地域です。
*   **インド、オーストラリア**は総売上および販売数量が低い傾向にあり、市場開拓やプロモーションが必要な地域として特定されました。
*   **日本、フランス、英国**は総売上が上位グループに属するものの、販売数量は中程度であり、高単価製品が売れている可能性が示唆されます。

#### 1.5 返品影響分析

*   全取引の**約15.5%（774件）が返品**されており、これによる金銭的損失は**約897,214.85**ドル（平均1,159.19ドル/件）に上ります。
*   **Electronics (17.99%)、Networking (16.72%)、Storage (15.36%)** のカテゴリーで返品率が高い傾向が見られました。これは高単価製品や複雑な製品が多いためと考えられます。
*   国別では、**英国 (17.86%)、フランス (17.73%)、日本 (15.87%)** で返品率が高いことが判明しました。これらは一部の主要収益地域と重なります。

4.  **インド、オーストラリアといった低パフォーマンス地域の市場開拓**に注力し、新たな収益源を確保する。
5.  顧客評価は売上に直接的な相関が低いものの、**高評価製品のブランド力向上と低・中評価製品の改善**は中長期的な顧客ロイヤルティとブランド価値向上に不可欠であり、継続的な投資が求められます。
